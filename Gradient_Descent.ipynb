```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
print (x)
```

```python
def loss(x):
    return (x-3)**2
```

```python
l = loss(x)
```

```python
plt.plot(x,l)
plt.show()
```

```python
w = 0.2
lr = 0.1

plt.plot(x,l)

for ix in range(20):
    plt.scatter(w,loss(w))
    w -= lr*2*(w-3)
    print ("W : {} , loss : {}".format(w,loss(w)))

plt.show()
```
