```python
import emoji
```

```python
emoji_dict = {
    "0" : ":heart:",
    "1" : ":baseball:",
    "2" : ":smile:",
    "3" : ":disappointed:",
    "4" : ":fork_and_knife:"
}
```

```python
def label_to_emoji(label):
    return emoji.emojize(emoji_dict[str(label)], use_aliases=True)
```

```python
for i in range(len(emoji_dict.keys())):
    print(label_to_emoji(i))
```
